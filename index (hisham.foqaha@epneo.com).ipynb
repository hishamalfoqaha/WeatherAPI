```python
import requests
import pandas as pd
from datetime import datetime

api_key = 'f1d7b7e9ce2bc4c7d82f2e73b500a8ba'
city_name = 'Dortmund'
start_date = datetime(2022, 1, 1)
end_date = datetime.now()
num_hours = int((end_date - start_date).total_seconds() / 3600)
num_6_hours = num_hours // 6
base_url = 'https://api.openweathermap.org/data/2.5/forecast'
params = {
    'q': city_name,
    'cnt': num_6_hours,  # Set cnt parameter to number of 6-hour intervals
    'start_date': int(start_date.timestamp()),
    'appid': api_key
}

# Create an API request
response = requests.get(base_url, params=params)
data = response.json()

#Filtering the results
weather_data = data['list']
weather_df = pd.DataFrame(weather_data)


weather_df = weather_df[['dt', 'main', 'weather']]
weather_df['date'] = pd.to_datetime(weather_df['dt'], unit='s')
weather_df['temperature'] = weather_df['main'].apply(lambda x: x['temp'] - 273.15)  # to make the temperatures in Celsius
weather_df['humidity'] = weather_df['main'].apply(lambda x: x['humidity'])
weather_df['weather_description'] = weather_df['weather'].apply(lambda x: x[0]['description'])

# Drop unwanted columns
weather_df = weather_df.drop(columns=['dt', 'main', 'weather'])
```

```python
import requests
import pandas as pd

# Set the API endpoint and parameters
url = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/Dortmund"
params = {
    "unitGroup": "metric",
    "include": "days",
    "key": "AKBLR5U2LLLMDZZF7YRCWZ2Y5",
    "contentType": "json"
}

# Send GET request to the API
response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the JSON data from the response
    weather_data = response.json()

    # Extract the relevant weather information
    city = weather_data["resolvedAddress"]
    current_temp = weather_data["currentConditions"]["temp"] if "currentConditions" in weather_data else None

    # Create a list of dictionaries to store daily weather data
    daily_weather_data = []
    for day in weather_data["days"]:
        daily_data = {
            "Date": day["datetime"],
            "Max Temp (°C)": day["tempmax"],
            "Min Temp (°C)": day["tempmin"],
            "Weather": day["conditions"]
        }
        daily_weather_data.append(daily_data)

    # Create a Pandas DataFrame from the list of dictionaries
    weather_df = pd.DataFrame(daily_weather_data)

    # Print the extracted weather information and display the DataFrame
    print("City:", city)
    print("Current temperature (°C):", current_temp)
    print("Weather forecast for the next 7 days:")
    print(weather_df)
else:
    print("Error occurred while retrieving weather data. Status code:", response.status_code)
```

```python
import requests
import pandas as pd

# Set the API endpoint and parameters
url = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/Dortmund?unitGroup=metric&include=hours&key=AKBLR5U2LLLMDZZF7YRCWZ2Y5&contentType=json"
params = {
    "startDateTime": "2021-01-01T00:00:00Z",
    "endDateTime": "2023-02-28T23:59:59Z"
}

# Send GET request to the API
response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the JSON data from the response
    weather_data = response.json()

    # Extract the relevant weather information
    city = weather_data["resolvedAddress"]
    current_temp = weather_data["currentConditions"]["temp"] if "currentConditions" in weather_data else None

    # Create a list of dictionaries to store hourly weather data
    hourly_weather_data = []
    for hour in weather_data["hours"]:
        hourly_data = {
            "Date": hour["datetime"],
            "Temperature (°C)": hour["temp"],
            "Weather": hour["conditions"]
        }
        hourly_weather_data.append(hourly_data)

    # Create a Pandas DataFrame from the list of dictionaries
    weather_df = pd.DataFrame(hourly_weather_data)

    # Print the extracted weather information and display the DataFrame
    print("City:", city)
    print("Current temperature (°C):", current_temp)
    print("Historical hourly weather data from January 2021 to February 2023:")
    print(weather_df)
else:
    print("Error occurred while retrieving weather data. Status code:", response.status_code)
```
